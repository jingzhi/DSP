# 1 Signals

## 1.1 definition of a signal

* a set of data or information
* a function  $ T \to A $ define on some time axis $ T $
  * $ T = \mathbb{R} $ : continuous time signal, $ f(k) $ denotes value of the signal at time k.
  * $ T = \mathbb{Z} $ : discrete time signal, $ f[k] $ denotes value of the signal at time k.
  * $ A = \mathbb{R} $ : real (valued) signal, we may allow $ A = \mathbb{R}^n $ for some positive integer n. 
  * $ A = \mathbb{C} $ : complex (valued) signal, we may allow $ A = \mathbb{C}^n $ for some positive integer n.

## 1.2 size of signals
Size indicates the strength of the signal, it must connote information about both the _amplitude_  and the _duration_ of the signal to give good indication.

Recall $$ |x(t)|^2 = 
\begin{cases}
x(t)x(t), \text{for real x(t)} \\
x(t)x^*(t), \text{for complex x(t)}
\end{cases} $$

1. signal energy
    $$ E_f = \int^{\infty}_{-\infty}|f(t)|^2 dt $$
    $$ E_f = \sum^{\infty}_{k=-\infty}|f[k]|^2 $$
    Signal energy doesn't indicate the actual energy of the signal, which depends on the load as well. Ex can be seen as resistor value normalised to one, and as indication of the energy capacity of the signal.

2. signal power
    $$ P_f =\frac{1}{T} \int^{\frac{T}{2}}_{-\frac{T}{2}}|f(t)|^2 dt $$
    $$ P_f =\frac{1}{T} \sum^{\frac{T}{2}}_{k=-\frac{T}{2}}|f[k]|^2 $$
    For aperiodic signal, $ T \to \infty $.
    Signal power is the time average of signal energy = mean square value of x(t)

## 1.5 signal operations

### sum of two signal (superposition)
g+f is the sum of signal g and signal f.
signal g+f is defined as 
$$ (f+g)(t) \triangleq f(t) +g (t) $$

### product of two signal (amplification)
g+f is the sum of signal g and signal f.
signal g+f is defined as 
$$ (f+g)(t) \triangleq f(t) +g (t) $$

### reserse signal
$$ f^r[k] \triangleq f[-k] $$
### conjugate signal
$$ f^c[k] \triangleq \overline{f[-k]} $$
note if f[k] is real valued, its reverse signal = its conjugate signal
### time shifting
Advance in time (shift left, +)
x(t) -> x(t+T)
// f(t-T) = x(t), f(t) = x(t+T), f takes value of x T earlier
Delay in time (shift right, -)
x(t) -> x(t-T)
// f(t+T) = x(t), f(t) = x(t-T), f takes value of x T later
Recall, shifting on an axis, + -> move in -ve direction, - -> move in +ve direction
### time scaling
Compress
x(t) -> x(ct)
Expand
x(t) -> x(t/c)
Recall, scaling of an axis, * -> dilact, / -> compress
### combined operation
Recall graph transformation, the order is important, each transformation operates on x alone.

# 2 Systems

## 2.1 definition of  a system

* processes signals to 
	1. modify them 
	2. extract information from them 
* comprises of two parts:
    1. a set of variables that take values in some configuration space
    2. the permitted behaviour of the system, which is a subset of the configuration space.
    
## 2.2 System classification
* __Dynamic system__: a system where some ot all of the variables ar functions of time
* __Continuous time (dynamic) system__: all time dependent variables are continuous-time signals
//note a continuous time system by defauly involves time hence must be a dynamic system
* __Discrete time (dynamic) system__: all time dependent variables are discrete-time signals
* <mark> __Def1.1__  __Linear system__ <mark/> : A system is linear if
    1. the configuration space is a vector space.
    2. the permitted behavior is a subspace of the configuration space.
       - Recall Subspace: 
            - contains origin
            - closed under addition and scalar multiplication
    3. <mark>__Prop1.1__ <mark/> Connecting two linear system yields a linear system.
    4. Example: sampling can be viewed as a linear system
    
* __Time-invariant (dynamic) system__: the permitted behaviour is closed under arbitary time shift.
* __Single-input single-output(SISO) system__: exactly one input and one output and no other terminals. (may have other variables)
* __Deterministic system__: if the output signal is fully determined by the input signal.
* __Causal system__: if at any time $t$, output $y(t)$ is fully determined by the values of input $u(t')$ with $t'\leq t$

## 2.3 SISO LTI system
Every LTI operation boils down to a convolution and vice versa.

The output of an LTI system is the convolution of the input signal with the imulse response of the system.

The LTI operation is especially useful since it allows us to track the transformation/process/system by tracking what happens to some basis. For which all inputs can be understood via linear combination of basis. We will be able to undertand the output of any input by understanding what happens to basis.

<mark>__prop1.2__<mark/> An LTI system is causal if and only if its impulse response is a causal signal.

<mark>__prop1.3__<mark/> An LTI system is BIBO stable if and only if its impulse response is a stable signal.( FT of impulse response (transfer function of the system) exists) (ROC of transfer function in laplace/z domain contains im axis/unit circle)

# 3 Convolution
## 3.1 Definition
Discrete time:
$$\begin{align}
(f*g)[n]&=\sum_{k\in\mathbb{z}}\sum_{l\in\mathbb{Z}}f[k]g[l]\delta[n-(k+l)]\\
&=\sum_{k\in\mathbb{Z}}f[k]g[n-k]\\
&=\sum_{k\in\mathbb{Z}}g[k]f[n-k]
\end{align}$$

Continuous time:
$$\begin{align}
(f*g)(t)&=\int_{-\infty}^{\infty}f(\tau)g(t-\tau)d\tau\\
&=\int_{-\infty}^{\infty}g(\tau)f(t-\tau)d\tau
\end{align}$$

If the sum/integral does not converge, the convolution is not defined.

## 3.3 Finite-sum approximation
<mark> __prop1.5__ Finite-sum approximation </mark>:

If:
* h[.] is stable
* u[.] is bounded with $|u[k]|\leq b\lt\infty$ 
* $y[.]=(u*h)[.]$
* $\tilde{y}_L[n]=\sum_{k:|k|\leq L}u[n-k]h[k]$

Then for every $\epsilon \gt 0$, there exists an integer $L_0 \lt \infty$ such that $|y|[n]-\tilde{y}_L[n]| \lt \epsilon b$

## 3.4 Associativity of convolution
<mark> __prop1.6__ Associativity of convolution </mark>

If:
* $f,g,h$ are all left-sided (or all right-sided)

Or:
* One of the signal is bounded, the other two are stable

Then
$$f*(g*h)=(f*g)*h$$
